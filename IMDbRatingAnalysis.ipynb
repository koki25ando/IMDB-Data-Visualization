```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go

rating = pd.read_csv('https://s3-ap-southeast-2.amazonaws.com/koki25ando/IMDb/ratings.csv', encoding='latin-1')
```

```python
release_year = rating['Year']
release_year_hist = go.Histogram(
    x = release_year.values,
    y = release_year.index,
    xbins = dict(
        size = 1
    )
)
data = [release_year_hist]
layout = go.Layout(
    title = 'Distribution of release year'
)
fig = go.Figure(data = data, layout = layout)
fig['layout']
py.iplot(fig)
```

```python
rating_scatter = go.Scatter(
    x = rating['IMDb Rating'],
    y = rating['Your Rating'],
    mode = 'markers',
    hoverinfo = 'text',
    text = rating['Title']
)
layout = go.Layout(
    title = 'IMDb Rating vs Your Rating',
    xaxis = dict(title = 'IMDb Rating'),
    yaxis = dict(title = 'Your Rating')
)
fig = go.Figure(data = [rating_scatter], layout = layout)
py.iplot(fig)
```

```python
director_bar = go.Bar(
    x = rating['Directors'].value_counts().index,
    y = rating['Directors'].value_counts().values
)
layout = go.Layout(
    title = 'Top Directors',
    xaxis = dict(range=[0, 20]),
    yaxis = dict(title = 'Number of Movies')
)

fig = go.Figure(data = [director_bar], layout = layout)
py.iplot(fig)
```
